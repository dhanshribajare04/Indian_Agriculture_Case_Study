```python
# Get unique values and their counts in the 'Season' column
season_counts = df['Season'].value_counts()
print(season_counts)

# Plot the distribution of the 'Season' column
plt.figure(figsize=(10, 6))
season_counts.plot(kind='bar')
plt.title('Distribution of Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

suggest the crop to be produce as per the season wise

```python
# Group by 'Season' and 'Crop' and sum the production
season_crop_production = df.groupby(['Season', 'Crop'])['Production'].sum().reset_index()

# Find the crop with the highest production for each season
recommended_crops = season_crop_production.loc[season_crop_production.groupby('Season')['Production'].idxmax()]

# Display the recommended crops for each season
recommended_crops
```

```python
x = recommended_crops['Season']
y = recommended_crops['Crop']
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Season')
plt.ylabel('Crop')
plt.title('Recommended Crops for Each Season')
plt.show()

# Display the recommended crops for each season
recommended_crops

```

```python
# Scatter plot to visualize the relationship between Area and Production
plt.figure(figsize=(10, 6))
plt.scatter(df['Area'], df['Production'], alpha=0.5)
plt.title('Relationship between Area and Production')
plt.xlabel('Area')
plt.ylabel('Production')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="Crop_Year", y="Production", hue="Crop", estimator="sum")
plt.title("Production Trend Over the Years")
plt.xticks(rotation=45)
plt.show()
```

```python
### 2️⃣ Top 5 Crops by Production ###
top_crops = df.groupby("Crop")["Production"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_crops.index, y=top_crops.values, palette="viridis")
plt.title("Top 5 Crops by Total Production")
plt.ylabel("Total Production")
plt.xticks(rotation=45)
plt.show()
```

```python
### 3️⃣ State-wise Production Distribution ###
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="State_Name", y="Production")
plt.xticks(rotation=90)
plt.title("State-wise Agricultural Production Distribution")
plt.show()
```

```python
### 4️⃣ Crop-wise Production Trend Over the Years ###
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Crop_Year", y="Production", hue="Crop", estimator="sum")
plt.title("Crop Production Trend Over the Years")
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
### 5️⃣ Correlation Between Area and Production ###
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Area", y="Production", alpha=0.5)
plt.title("Correlation Between Area and Production")
plt.xlabel("Area (in hectares)")
plt.ylabel("Production")
plt.show()
```

```python
# Correlation coefficient
correlation = df[["Area", "Production"]].corr().iloc[0, 1]
print(f"Correlation between Area and Production: {correlation:.2f}")
```

1️⃣ Overall Agricultural Production Trend Over the Years
The total agricultural production has fluctuated over the years, with certain periods showing growth while others experience a decline.
Possible reasons: climate changes, government policies, natural calamities, or technological advancements in farming.

2️⃣ Top 5 Crops by Total Production
The most-produced crops in India (based on total production) include Rice, Wheat, Sugarcane, Maize, and Cotton.
These crops are staple food sources and contribute significantly to India's economy and food security.

3️⃣ State-wise Production Distribution
Some states exhibit higher production levels, likely due to fertile land, irrigation facilities, and government policies supporting agriculture.
States with lower production might be affected by droughts, lack of resources, or smaller agricultural land availability.

4️⃣ Crop-Wise Production Trend Over the Years
Certain crops, such as Rice and Wheat, show consistent growth over the years, reflecting stable demand and improved farming techniques.
Other crops exhibit fluctuations, which may be due to changing climate conditions, market demand, or soil fertility.

5️⃣ Correlation Between Area and Production
There is a positive correlation between Area and Production, meaning that larger land areas typically result in higher production.
However, some data points may indicate efficient farming techniques (high yield in small areas) or land underutilization.
